```python
import requests
from bs4 import BeautifulSoup
```

```python
# Set start and end dates
start_date = datetime.datetime(2022, 1, 1)
end_date = datetime.datetime(2022, 12, 31)

# Define function to get news for a given date range
def get_news(start, end):
    news = finnhub_client.company_news('AAPL', _from=start.strftime('%Y-%m-%d'), to=end.strftime('%Y-%m-%d'))
    return news

# Define function to scrape full article text
def get_full_text(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    article = soup.find('article')
    if article:
        return article.text.strip()
    else:
        return ''

# Make requests in a loop
news_list = []
while end_date >= start_date:
    print(f"Requesting news for {start_date.date()} to {end_date.date()}")
    news = get_news(start_date, end_date)
    for article in news:
        url = article.get('url')
        if url:
            full_text = get_full_text(url)
            article['full_text'] = full_text  # Add full text to original dict
        else:
            article['full_text'] = ''
        news_list.append(article)
    end_date -= datetime.timedelta(days=8)

# Create DataFrame
df = pd.DataFrame(news_list)

# Print summary
print(f"Got {len(df)} news articles for AAPL in 2022.")
```

```python
import requests
import datetime

# Set API endpoint and parameters
url = "https://www.alphavantage.co/query"
params = {
    "function": "NEWS_SENTIMENT",
    "tickers": "AAPL",
    "apikey": "GXRN5NFIZKLQXBUQ",
    "limit": "200"  # Set page size to 200 (maximum allowed)
}

# Set date range
start_date = datetime.datetime(2022, 1, 1)
end_date = datetime.datetime(2022, 12, 31)

# Initialize variables
all_articles = []
done = False

# Loop through pages of API results
while not done:
    # Add time_from parameter to API parameters
    params["time_from"] = start_date.strftime("%Y%m%dT%H%M")
    
    # Make request to API
    response = requests.get(url, params=params)
    data = response.json()
    
    # Check if no more results
    if not data["feed"]:
        done = True
    else:
        # Loop through articles in page and check date
        for article in data["feed"]:
            timestamp_str = article['time_published']
            timestamp_utc = datetime.datetime.strptime(timestamp_str, '%Y%m%dT%H%M%S')
            if start_date <= timestamp_utc <= end_date:
                all_articles.append(article)
        
        # Check if reached end date
        if timestamp_utc < start_date:
            done = True
        else:
            # Increment start date
            start_date = timestamp_utc + datetime.timedelta(seconds=1)

# Print all articles
for article in all_articles:
    print(article)
```
