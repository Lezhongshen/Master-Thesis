```python
result = []
```

```python
for i, tweet in enumerate (sntwitter.TwitterSearchScraper(keyword + " since:" + start_dt + " until:" + end_dt).get_items()):
    try:
        lan=tweet.lang
        #if i > num_tweets_to_scrape:
        #  break
        if lan in languages:
          result.append(tweet)
          print(tweet.date)
          tweet.date = tweet.date.dt.tz_localize(None)
          #ws.append([tweet.id, tweet.date, tweet.rawContent])
          
        else:
          pass
    except:
        lan='error'
#wb.save("drive/My Drive/WMTTweet.xlsx")
        
```

```python
# Extracting the desired information into a DataFrame
df = pd.DataFrame([[tweet.id, tweet.date, tweet.rawContent] for tweet in result], columns=['id', 'date', 'rawContent'])
```
