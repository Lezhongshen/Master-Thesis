```python
import requests
import re
from bs4 import BeautifulSoup
from nltk.tokenize import sent_tokenize, word_tokenize
from tqdm import tqdm
```

```python
related_text = []
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}


for x in tqdm(df['url'], desc='Scraping News Articles'):

# Make a GET request to the URL
  response = requests.get(x, headers=headers)

# Use BeautifulSoup to parse the HTML content of the response
  soup = BeautifulSoup(response.content, 'html.parser')

# Extract the text from the HTML content and split it into sentences
  text = soup.get_text()
  sentences = sent_tokenize(text)

  aapl_sentences = []
  try:
    for sentence in sentences:
      if re.search(r'\bApple\b', sentence):
          # Tokenize the sentence into words
          words = word_tokenize(sentence)

          if 'Apple' in words:
            # Find the index of the first occurrence of 'Apple'
            index = words.index('Apple')
            # Extract the 20 words before and after 'Apple'
            start = max(0, index - 20)
            end = min(len(words), index + 21)
            aapl_context = ' '.join(words[start:end])
            aapl_sentences.append(aapl_context)

    related_text.append(aapl_sentences)
  except ValueError:
          related_text.append('')
```
