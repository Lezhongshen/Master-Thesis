```python
df_tweet1 = df_tweets[['date','rawContent','text_preprocess_withemo','sentiment_prediction','stock_symbol']].copy()
```

```python
# convert the 'datetime' column to a datetime dtype
df_tweet1['date'] = pd.to_datetime(df_tweet1['date'])

# apply the date transformation based on the time
df_tweet1['Date'] = df_tweet1['date'].apply(lambda x: x.date() if x.time() >= pd.to_datetime('09:30:00').time() else (x.date() + pd.Timedelta(days=-1)))

# print the resulting dataframe
df_tweet1
```

```python
df_tweet2 = df_tweet1[['Date','rawContent','text_preprocess_withemo','sentiment_prediction','stock_symbol']].copy()
df_tweet2
```

```python
# convert date_column to datetime
#df_tweet1['date'] = pd.to_datetime(df_tweet1['date']).dt.date

# sort the dataframe by the date_column in ascending order
df_tweet2.sort_values(by='Date', inplace=True)
```

```python
df_tweet2 = df_tweet2.reset_index(drop=True)
```

```python
df_tweet2
```

```python
# convert string '2022-03*-01' to datetime.date object
date_str = '2022-03-01'
date = datetime.strptime(date_str, '%Y-%m-%d').date()
```

```python
df_tweet_final = df_tweet2[df_tweet2['Date'] >= date]
```

```python
df_tweet_final = df_tweet_final[df_tweet_final.stock_symbol != 'MRK']
```

```python
df_tweet_final
```

```python
# group by date and calculate count and mean of sentiment
result = df_tweet_final.groupby(['stock_symbol','Date'])['sentiment_prediction'].value_counts().unstack(fill_value=0)

# calculate the mean sentiment for each date
result['sentiment_mean'] = df_tweet_final.groupby(['stock_symbol','Date'])['sentiment_prediction'].mean()

# rename columns
result.columns = ['tweet_sentiment_count_0', 'tweet_sentiment_count_1', 'tweet_sentiment_count_2', 'tweet_sentiment_mean']

# print the result
result
```

```python
df = pd.read_csv('drive/My Drive/thesis/dataset/final_dataset_with_all_stock.csv [Final one]')
```

```python
df = df.drop(df.columns[0],axis=1)
```

```python
df
```

```python
df.columns
```

```python
df = df.rename(columns={'stocktwit_sentiment_count_0': 'stocktwit_sentiment_count_1', 'stocktwit_sentiment_count_1': 'stocktwit_sentiment_count_0'})
```

```python
df = df.drop(['stocktwit_sentiment_mean'],axis=1)
```

```python
df.columns
```

```python
df['stocktwit_sentiment_mean'] = (df['stocktwit_sentiment_count_0']*0 + df['stocktwit_sentiment_count_1']*1) / (df['stocktwit_sentiment_count_0'] + df['stocktwit_sentiment_count_1'])
```

```python
df.columns
```

```python
df_final = df[['stock_symbol', 'Date', 'Open', 'Close', 'High', 'Low', 'Volume',
       'macd', 'tweet_sentiment_count_0', 'tweet_sentiment_count_1',
       'tweet_sentiment_count_2', 'tweet_sentiment_mean',
       'stocktwit_sentiment_count_0', 'stocktwit_sentiment_count_1', 'stocktwit_sentiment_mean',
       'news_sentiment_count_0', 'news_sentiment_count_1',
       'news_sentiment_count_2', 'news_sentiment_mean', 'trends',
       'Null_value_news', 'Null_value_stocktwit', 'week_day', 'Thursday',
       'Tuesday', 'Wednesday']].copy()
```

```python
df_final['tweet_0_pct'] = df_final['tweet_sentiment_count_0'] / (df_final['tweet_sentiment_count_0'] + df_final['tweet_sentiment_count_1'] + df_final['tweet_sentiment_count_2'])
df_final['tweet_1_pct'] = df_final['tweet_sentiment_count_1'] / (df_final['tweet_sentiment_count_0'] + df_final['tweet_sentiment_count_1'] + df_final['tweet_sentiment_count_2'])
df_final['tweet_2_pct'] = df_final['tweet_sentiment_count_2'] / (df_final['tweet_sentiment_count_0'] + df_final['tweet_sentiment_count_1'] + df_final['tweet_sentiment_count_2'])
```

```python
df_final['stocktwit_0_pct'] = df_final['stocktwit_sentiment_count_0'] / (df_final['stocktwit_sentiment_count_0'] + df_final['stocktwit_sentiment_count_1'])
df_final['stocktwit_1_pct'] = df_final['stocktwit_sentiment_count_1'] / (df_final['stocktwit_sentiment_count_0'] + df_final['stocktwit_sentiment_count_1'])
```

```python
df_final['news_0_pct'] = df_final['news_sentiment_count_0'] / (df_final['news_sentiment_count_0'] + df_final['news_sentiment_count_1'] + df_final['news_sentiment_count_2'])
df_final['news_1_pct'] = df_final['news_sentiment_count_1'] / (df_final['news_sentiment_count_0'] + df_final['news_sentiment_count_1'] + df_final['news_sentiment_count_2'])
df_final['news_2_pct'] = df_final['news_sentiment_count_2'] / (df_final['news_sentiment_count_0'] + df_final['news_sentiment_count_1'] + df_final['news_sentiment_count_2'])
```

```python
df_final
```

```python
df_final.loc[df_final["stocktwit_sentiment_mean"].isnull(), "stocktwit_sentiment_mean"] = 0.5
```

```python
df_final.loc[df_final["stocktwit_0_pct"].isnull(), "stocktwit_0_pct"] = 0
df_final.loc[df_final["stocktwit_1_pct"].isnull(), "stocktwit_1_pct"] = 0
df_final.loc[df_final["news_0_pct"].isnull(), "news_0_pct"] = 0
df_final.loc[df_final["news_1_pct"].isnull(), "news_1_pct"] = 0
df_final.loc[df_final["news_2_pct"].isnull(), "news_2_pct"] = 0
```

```python
df_final.to_csv('drive/My Drive/thesis/dataset/final_dataset_with_all_stock.csv [Final one]')
```
